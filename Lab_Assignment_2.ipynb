```python
import urllib.request

# Step 1: Provide the URL of the image
image_url = 'https://w0.peakpx.com/wallpaper/364/878/HD-wallpaper-nature-beautiful-mor-pankh-mor-pankh.jpg'

# Step 2: Download the image using urllib, with custom headers
try:
    # Create a request object with a custom User-Agent
    req = urllib.request.Request(image_url, headers={'User-Agent': 'Mozilla/5.0'})
    with urllib.request.urlopen(req) as response:
        img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)
except urllib.error.HTTPError as e:
    print(f"Error downloading image: {e}")
    print("Possible reasons: Website restrictions, blocked User-Agent, missing permissions.")
    print("Try using a different image URL or checking the website's terms of service.")
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

# Load an image, flip it upside down, rotate it 180 clockwise and counterclockwise, and display all versions

```python
# Load the image
pic = cv2.imread("/content/drive/MyDrive/Digital Image processing/HD-wallpaper-nature-beautiful-mor-pankh-mor-pankh.jpg")

# Convert from BGR to RGB (Matplotlib expects RGB images)
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)

# Flip and rotate the image
up = cv2.flip(pic, -1)  # Flip both vertically and horizontally
# Use cv2.ROTATE_180 instead of cv2.ROTATE_180_CLOCKWISE for older OpenCV versions
left = cv2.rotate(pic, cv2.ROTATE_180)  # Rotate 180 clockwise
right = cv2.rotate(pic, cv2.ROTATE_180)  # Rotate 180 counterclockwise (same as clockwise for 180)


# Function to display images using Matplotlib
def show_image(img, title):
    plt.imshow(img)
    plt.axis('off')  # Remove axes for a cleaner display
    plt.title(title)
    plt.show()

# Display images
show_image(pic, "Original Image")
show_image(up, "Flipped Image (Upside Down)")
show_image(left, "Rotated 180 Clockwise")
show_image(right, "Rotated 180 Counterclockwise")
```
