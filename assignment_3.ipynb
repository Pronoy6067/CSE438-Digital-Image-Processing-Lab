```python
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image2 = np.ones((400, 400,3 ), dtype=np.uint8)*255
cv2.circle(image2, (200, 200),100, (255, 0, 255,), -1)
cv2_imshow(image2)
```

```python
image1 = np.ones((400, 400, 3), dtype=np.uint8)*255
cv2.rectangle(image1, (50, 100), (300, 350), (0, 255, 0), -1)
cv2_imshow(image1)
```

```python
addition = cv2.add(image1, image2)
cv2_imshow(addition)
```

```python
subtraction = cv2.subtract(image1, image2)
cv2_imshow(subtraction)
```

```python
multipliction = cv2.multiply(image1, image2)
cv2_imshow(multipliction)
```

```python
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray1)
cv2_imshow(gray2)
```

```python
division = cv2.divide(gray1, gray2, dtype=cv2.CV_32F)
division = cv2.normalize(division, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
cv2_imshow(division)
```

```python
bitwise_and = cv2.bitwise_and(gray1, gray2)
cv2_imshow(bitwise_and)
```

```python
bitwise_or = cv2.bitwise_or(gray1, gray2)
cv2_imshow(bitwise_or)
```

```python
bitwise_not1 = cv2.bitwise_not(gray1)
cv2_imshow(bitwise_not1)
```
