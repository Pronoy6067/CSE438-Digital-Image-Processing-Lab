## 1. b)	Write a program to read, write and display an image. [From URL]

```python
# Step 1: Provide the URL of the image
image_url = 'https://images.unsplash.com/photo-1461749280684-dccba630e2f6?q=80&w=1469&auto=format&fit=crop&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D'  # Replace with the actual URL
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

```python
#step 1
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(f"Shape of the Graycale Image : {gray_image.shape}")

plt.figure(figsize=(6,6))
plt.title("Grayscale Image")
plt.imshow(gray_image, cmap='gray') # Changed 'camp' to 'cmap'
plt.axis('off') # Changed plt('off') to plt.axis('off')
plt.show()
```

```python
img_crop=img_rgb[0:400,0:500] # Changed 'image_rgb' to 'img_rgb'

plt.imshow(img_crop) # Changed 'plt.show' to 'plt.imshow' to display the image
plt.axis('off')
plt.title("cropped Image")
plt.show()

```
